```python
#column 'MntGoldProds', follows an exponential distribution
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import expon

# Plot histogram
sns.histplot(s_df['MntGoldProds'], bins=20, kde=False, color='blue')
plt.title('Histogram of MntGoldProds')
plt.xlabel('Amount Spent on Gold Products')
plt.ylabel('Frequency')
```

```python
# Fit the data to an exponential distribution and plot
params_exp = expon.fit(s_df['MntGoldProds'])
x = np.linspace(0, s_df['MntGoldProds'].max(), 100)
pdf_exp = expon.pdf(x, *params_exp)
plt.plot(x, pdf_exp * len(s_df['MntGoldProds']), color='blue', label='Exponential Fit')

plt.legend()
plt.show()

print(f'Parameters for Exponential distribution: {params_exp}')
print(f'Probablity density function: {pdf_exp}')
print(f'cummulative distribution function: {cdf_exp}')
```

```python
cdf_exp = expon.cdf(x, *params_exp)
plt.plot(x, cdf_exp * len(s_df['MntGoldProds']), color='blue', label='Exponential Fit')
print(f'cummulative distribution function: {cdf_exp}')
```

```python
# checking if 'Income'  follow a Binomial distribution
from scipy.stats import binom
sns.histplot(s_df['Income'], bins=20, kde=False, color='blue')
plt.title('Histogram of income')
plt.xlabel('customers yearly household income')
plt.ylabel('Frequency')
plt.show()
```

```python
#Calculate probabilities for Binomial distribution
prob_binomial = binom.pmf(k=2, n=10, p=0.2)  # Probability of getting exactly 2 successes in 10 trials with 0.2 probability of success
expected_value_binomial = binom.mean(n=10, p=0.2)  # Expected value for Binomial distribution

print(f'Probability for Binomial distribution: {prob_binomial}')
print(f'Expected value for Binomial distribution: {expected_value_binomial}')
```
